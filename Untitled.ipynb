```python
# arrangement is [stafford, graduate PLUS]
import numpy as np
import matplotlib.pyplot as plt
interest = np.array([0.0608, 0.0708])
monthly_interest = 1+interest/12
principal = np.array([25000+23000+21000, 62000+60000+54000])
monthly = np.array([2000, 4000, 6000, 8000, 10000])
current = [principal.copy() for i in range(len(monthly))]
result = [[np.sum(current[0])] for i in range(len(monthly))]
```

```python
i=0
while (np.max(np.sum(current, axis=1)))>0:
    i+=1
    #compounding step
    current *= monthly_interest
    #payoff step
    for s, payment in enumerate(monthly):
        if np.sum(current[s]) <= 0:
            continue
        elif current[s][1] > 0:
            if current[s][1] >= payment:
                current[s][1] -= payment
            else:
                current[s][0] -= (payment - current[s][1])
                current[s][1] = 0
        elif current[s][0] >= payment:
            current[s][0] -= payment
        else:
            current[s][0] = 0
        result[s].append(np.sum(current[s]))
```

```python
for i, res in enumerate(result):
    plt.plot(res, label=f'${monthly[i]}/month')
plt.legend()
plt.xticks([i*12 for i in range(20)])
plt.gca().set_xticklabels([f'{i}' for i in range(20)]);
plt.xlabel('years')
plt.ylabel('remaining balance')
plt.yticks([i*50000 for i in range(6)])
plt.gca().set_yticklabels([f'${i*50}k' for i in range(6)])
```
